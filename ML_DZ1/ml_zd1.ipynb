# Cамый ветрянный месяц, самый влажный год  и неделя с самой переменчивой погодой

```
weather_hist['Year'] = weather_hist.date.dt.year
weather_hist['Month'] = weather_hist.date.dt.month
weather_hist['Week'] = weather_hist.date.dt.week 

weather_hist.head(5)
```

```
years = weather_hist.groupby('Year')

#years.head(5)
```

```
monthes = weather_hist.groupby('Month')


#monthes.head(5)
```

```
weeks = weather_hist.groupby('Week')

#weeks.head(5)
```

```
monthes.dtypes
```

```
monthes.sum(level='wind_speed')
```

```
monthes.sum(level='wind_speed').agg(['max'])
```

Сопоставив таблицы, делаем вывод что март - самый ветрянный месяц.
